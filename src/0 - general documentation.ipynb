<h1 id='0-problem-description' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>📝 | Problem Description</h1>

Recommendation Systems are present everyday in our lives helping us to find the perfect anime to watch on Crunchyroll, to keep us binge watching related YouTube videos or even to buy new products on Amazon. I can spend the whole day listing examples about these systems, but I think that you got the glimpse where Recommendation Systems are used, don't you?

However, have you ever get bored on Netflix and Instagram because they only recommend items related to the same topic? It is not because I liked two videos about cars on Instagram that I am a huge fan of cars and want to see only reels and posts about it. Or it is not because I saw a gamer chair price on Amazon that I want to buy only gamer chairs. When you keep receiving recommendations only related to a especific topic or product, we say that you are into a Limited Bubble of Recommendations.

Limited Bubble of Recomendations tends to make companies lose customers, so, in order to minimize these bubbles and to expand the recommendations' diversity, algorithms have been tuned and new variables have been considered to the model.

Nowadays, there are four main Recommendation Algorithms: `Demographic Filtering, Content-Based Filtering, Collaborative Filtering and Hybrid Filtering`, let's see how each one works assuming that we are dealing with Animes Recommendation.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Demographic Filtering</p>

> **Logic** - `it recommends animes that are popular between all users. Normally recommends the animes with higher Bayesian Mean Score, Arithmetic Mean Score and Popularity Rank`;

> **Example** - `if Bleach, Oregairu and Jujutsu Kaisen are the animes with higher scores on a stream platform, this platform will recommend them to all users, regardless their tastes`;

> **Metrics** - `Arithmetic Mean Score (when dealing with few ratings data), Bayesian Mean Score (when dealing with many ratings data) and Popularity Rank (when dealing with only popularity and not ratings)`;

> **Advantages** - `1) only animes data is required; 2) the model training takes less time to compute; 3) the metric is not complicated to understand`;

> **Disadvantages** - `1) bad recommendations are made since this technique generalizes the taste of all users to only one: the animes with higher Score and Popularity`.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Content-Based Filtering</p>

> **Logic** - `it recommends animes that are similar to other animes that a user liked. There are two variations of this technique: Plot Description Based (it recommends animes with similar synopsis) and Metadata Based (it recommends animes with similar metadata, such as genres, licensors, authors, producers and studios)`;

> **Example** - `if a user liked Dragon Ball Z, there is a high probability that the platform will recommend Dragon Ball Super or Naruto (both shounen animes with superpowers)`;

> **Metrics** - `Cosine Similarity Score, Jaccard Simiarity and Pearson Correlation. In this project, we will be working with Cosine Similarity Score`;

> **Advantages** - `1) better recommendations are provided compared to Demographic Filtering Technique; 2) it recommends similar items that the users liked; 3) each user receives personalized recommendations`;

> **Disadvantages** - `1) more data is required, such as items' synopsis and metadata; 2) when working on Plot Description Based Model, there is a probability to get only sequels and prequels recommendations; 3) when working on Metadata Based Model, there is a probability to create a Bubble that only recommends animes with a specific genre`.

<br /><br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Collaborative Filtering</p>

> **Logic** - `it recommends animes that are similar to animes liked by similar users. There are two variations of this technique: User Based Filtering (when dealing with users similarity) and Item Based Filtering (when dealing with items similarity). Both of them follow the Memory/Neighborhood Logic`;

> **Example** - `if user A and user B liked Demon Slayer, user A is considered similar to user B. If user B also liked Grand Blue, the platform will recommend Grand Blue to user A`;

> **Metrics** - `Cosine Similarity Score, Jaccard Similarity, Pearson Correlation and Root Mean Squared Error. In this project, we will be working with Pearson Correlation and Root Mean Squared Error`;

> **Advantages** - `1) better recommendations are provided compared to Content-Based and Demographic Filtering; 2) it recommends similar items that similar users liked; 3) each user receives personalized recommendations; 4) there is a few probability to create a Bubble of Contents`;

> **Disadvantages** - `1) anime and user data are required; 2) the dataset must contain a significant amount of user ratings to animes; 3) there is yet a probability of creating a Bubble of Contents; 4) it is needed to add a cut-off amount of ratings and mean rating score by user in order to avoid outliers; 5) the algorithm is more complex and takes more time to run`.

<br /><br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Hybrid Filtering</p>

> **Logic** - `it combines the Content-Based and Collaborative Filtering techniques. It normally applies the first one when there are few users ratings available and smoothly replaces it to the second one as more users ratings become available`;

> **Example** - `1) When few users ratings are availabe: if a user liked Dragon Ball Z, there is a high probability that the platform will recommend Dragon Ball Super or Naruto (both shounen animes with superpowers); 2) When many users ratings are available: if user A and user B liked Demon Slayer, user A is considered similar to user B and user B also liked Grand Blue, the platform will recommend Grand Blue to user A`;

> **Metrics** - `Cosine Similarity, Pearson Correlation, Jaccard Similarity, Root Mean Squared Error, ROC Curve, and so on. In this project, we will be working with Cosine Similarity, Pearson Correlation and Root Mean Squared Error`;

> **Advantages** - `1) since it combines two techniques, it contains the advantages related to both of them; 2) also, it decreases the probability of making poor recommendations when the platform is new and does not contain many users ratings`;

> **Disadvantages** - `1) since it combines two techniques, it contains the disadvantages related to both of them; 2) also, it requires more computational resources since the algorithm is more complex and complete than the previous ones`.

<h1 id='4-metrics' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>📏 | Metrics</h1>

Four metrics were used to evaluate the Models: `Bayesian Mean` for Demographic Filtering, `Cosine Similarity` for Content-Based Filtering, `Pearson Correlation` and `Root Mean Squared Error [RMSE]` for Collaborative Filtering.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Bayesian Mean</p>

When working with mean scores of products, we have a tendency to use the `Arithmetic Mean` rather than other metrics, however, it can be a problem for Recommendation System Models due to the difference between the score value and the amount of scores. For instance, picture an anime that has a Mean Score of 9.8 and only 100 votes, and another anime with a Mean Score of 8.5 and 100,000 votes. Which one is better? The one with the higher score and less votes or the one with the smaller score and more votes? In this case and statistically speaking, it is correct to assume that the one with the smaller score and more votes is the better because the difference of the amount of votes between the two animes is huge. So, we have just stumbled upon a problem: there are situations that the Arithmetic Mean is not suitable and can compromisse Recommendation Systems Models.

Having this in mind, we can use the `Cumulative Mean`. Its idea is similar to the Arithmetic one, but the difference is that the value is not divided by the number of votes. So, if an anime got two ratings of 10, we say that its Cumulative Mean is 20, while its Arithmetic Mean is 10. But, this method has a problem too: consider an anime with 1,000,000 ratings of 1 score and another anime with 10,000 ratings of 10 score, if we calculate their Cumulative Mean, the first anime will have a score of 1,000,000 whereas the second one will have a score of 100,000. So, even though having the better ratings, the anime can be in a bad rank position if the number of votes is substancial lower than an anime with worse ratings and a huge number of votes.

So, up to now, we have this two metrics problems:

> **Arithmetic Mean** - `does not assume that an anime with lower score and huge number of ratings is better than an anime with higher score and only a few number of ratings`;

> **Cumulative Mean** - `does not assume that an anime with a lower score and huge number of good ratings is better than an anime with higher score and a more huge number of bad ratings`.

<br /><br />

In order to solve these two problems and to giver better, realistic mean scores, the `Bayesian Mean` has been created!! Its idea is simple, first we have to find the number of ratings of the item, the Arithmetic Mean of the item, the minimum number of ratings that an item must have to be considered in the system and the Arithmetic Mean across all items. After that, we apply them in the following equation:

```python
bayeasian_mean = ((v / (v + m)) * A) + ((m / (v + m)) * C)
```

$$
Bayesian Mean = \frac{v}{v + m} \cdot A + \frac{m}{v + m} \cdot C
$$

where:

- v: number of the item's ratings;
- m: minimum number of ratings that an item must have to be considered in the system;
- A: item's Arithmetic Man;
- C: Arithmetic Mean across all items.

<br />

The equation can also be used in its alternative form, resulting in the same result:

```python
bayesian_mean = ((c * m) + (n * r)) / (c + n)
```

$$
Bayesian Mean = \frac{(c \cdot m) + (n \cdot r)}{(c + n)}
$$

where:

- c: minimum number of ratings that an item must have to be considered in the system;
- m: Arithmetic Mean across all items;
- n: number of the item's ratings;
- r: item's Arithmetic Mean.

<br />

Due to its good results, Bayesian Mean is used by IMDB to calculate the mean score of movies listed on the website.

Also, it is highly recommended to start using Arithmetic Mean when there are a few ratings to an item and smoothly migrate to Bayesian Mean accordingly the number of ratings increase.

Some observations about this metric:

- for an item with a fewer than average number of ratings - the score should be around the system’s Arithmetic Mean;
- for an item with a substantial number of ratings - the score should be the item’s Arithmetic Mean;
- as the number of ratings that an item receives increases, the score should gradually move from system’s Mean to item’s Mean.

<br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Cosine Similarity</p>

`Cosine Similarity` is a perfect metric to measure the similarity between two values of a same feature, especially when dealing with sequential texts or rating scores. In a nutshell, it works dividing the sum of the multiplication of the appearance of the elements in each items by the product of the squared root of the quadratic appearance of each element into the items 🤯😶‍🌫️. You can be wondering: *What the hell did I just read?*. Do not worry, The example below makes everything clear 🤣

Consider these two sentences: `Hello World!` (sentence A) and `Hello!` (sentence B). The first thing to do is to create a table showing the frequency of each word in each sentence:

After that, we: 

- plot the frequencies using points;
- trace two lines across the points, one line to one point; 
- calculate the angle between the lines;
- calculate the cosine of the angle;
- the cosine of the angle is the Cosine Similarity Score. It goes from 0 (completely not similar) to 1 (completely similar).

Now that we already know how it works, let's take a peek at its equation:

```python
sum(Wq[k] * Wd[k]) / (sqrt(sum(Wq[k]**2)) * sqrt(sum(Wd[k]**2)))
```

$$
Cosine Similarity = \frac{\sum_{k=1}^{t} (Wq[k] \cdot Wd[k])} {\sqrt{\sum_{k=1}^{t} (Wq[k]^2)} \cdot \sqrt{\sum_{k=1}^{t} (Wd[k]^2)}}
$$

where:

- Wq and Wd: items;
- k: each word present into the Bag of Words;
- t: amount of words into the Bag of Words;
- Wq[k] and Wd[k]: frequency of 'k' word into 'Wq' and 'Wd' items.

<br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Pearson Correlation</p>

For Collaborative Filtering, Pearson Correlation is better suited than Cosine Similarity as our metric. Here you go some reasons:

> **Sparse Data** - `Pearson Correlation is better than Cosine Similarity in both Collaborative and Hybrid Filtering due to the input data be sparse. Pearson Correlation also excludes null/nan/sparse registrations in the calculations whereas Cosine Similarity considers as 0 (which is not appropriate)`;

> **Vectors Magnitudes** - `Pearson Correlation considers the magnitude (strong) of the vectors, whereas Cosine Similarity only considers the direction. Like, in Content-Based Filtering, if a word appears three times, it'll have the weight than a word that appears only once when using Cosine Similarity. In the other hand, the word with more frequency will have more weight when applying Pearson Correlation`.

<br />

To calculate the Pearson Correlation, we have to:

- Plot one variable on X-axis and another variable on Y-axis;
- Plot points on (X_variable_value, Y_variable_value) coordenates;
- Calculate the `Linear Regression` between them and trace the Regression Line;

While when the Correlation Value is positive, we say that the higher the variable A is, the higher the variable B is too and that there is a `Positive Covariance` between them;

Equation:

```python
pearson_correlation = covariance(x, y) / (standard_deviation(x) * standard_deviation(y))
```

$$
\text{pearson_correlation} = \frac{covariance(x, y)}{\sigma(x) \cdot \sigma(y)}
$$

<br />

where:

- x: first variable;
- y: second variable.

<br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Root Mean Squared Error [RMSE]</p>

`Root Mean Squared Error [RMSE]` is a metric used to measure the accuracy of Machine Learning Models' predictions and we will be using it to evaluate the predicted users ratings!!

```python
rmse = sqrt(sum((yi - yp)**2) / n)
```

$$
\text{RMSE} = \sqrt{\frac{\sum_{i=1}^{n}{(y - y')^2}} {n}}
$$

where:

- n: number of predicted ratings;
- y: real ratings scores;
- y': predicted ratings scores.
